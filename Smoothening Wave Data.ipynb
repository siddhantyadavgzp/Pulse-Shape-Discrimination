```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np
from scipy.signal import savgol_filter

df= pd.read_csv('OneDrive/Documents/Karm/ANeutron.csv',header=None)

df.shape
```

```python
matrix=df.values
```

```python
with open('OneDrive/Documents/Karm/FNeutron.txt', 'a') as f:
    k=0
    for i in matrix:
        k=k+1
        j = np.round(savgol_filter( i, window_length=7, polyorder=2),1)
        np.savetxt(  f, [j], delimiter=' ', fmt='%s')
        if k==50000:
            break;
```

```python
import matplotlib.pyplot as plt
x=np.arange(191)
k=10
plt.plot(x, matrix[k], label='Noisy Signal')
plt.plot(x,np.round(savgol_filter( matrix[k], window_length=7, polyorder=2),1),label='Filtered' )
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.legend()
plt.show()
```
