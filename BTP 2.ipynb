```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
i=100
j=50
app=[]
while i < 1000:
    fdf = pd.DataFrame(df)
    ffdf = fdf[(fdf['peak']>=i) & (fdf['peak']<i+j)]
    k = 2  # Number of clusters
    kmeans = KMeans( n_clusters = k , init = 'k-means++' )
    kmeans.fit(fdf[['cr']])
    app.append(kmeans.labels_)
    j*=2;
    i+=j;
app
```
